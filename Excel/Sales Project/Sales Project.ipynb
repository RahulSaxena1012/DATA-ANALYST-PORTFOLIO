```python
#Importing all the necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_df = pd.read_csv("salesdata_original.csv", parse_dates=["Order Date"], dayfirst=True)
```

```python
sales_df.info()
```

```python
sales_df.sort_values("Order Date", ascending= False)
```

```python
sales_df = sales_df.assign(
Year=sales_df["Order Date"].dt.year,
Month = sales_df["Order Date"].dt.month
)
```

```python
sales_df
```

```python
sales_df_pivot = sales_df.pivot_table(
                index="Sub-Category",
                values= "Sales",
                aggfunc="sum"
                )
```

```python
sales_df.to_excel("SalesData.xlsx", sheet_name="Sales Data", index=False)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

sns.barplot(x=sales_df_pivot.index,
       y="Sales",
       data=sales_df_pivot)

plt.xticks(rotation=90)
```
